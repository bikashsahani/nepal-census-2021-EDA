5. What is the range and distribution of Total Family Number and Total Household Number?

```python
# to find range 
min_family_num = data['Total family number'].min()
max_family_num = data['Total family number'].max()
family_range = max_family_num - min_family_num
print(f"Range of Total family number = {family_range}")

min_household = data['Total household number'].min()
max_household = data['Total household number'].max()
household_range = max_household - min_household
print(f"Range of Total household number = {household_range}")
```
