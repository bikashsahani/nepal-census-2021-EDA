9. Identify and handle any outliers in Total Population or Household counts.

```python
# outliers for Total population
# Calculate Q1 (25th percentile) and Q3 (75th percentile)
population_q1 = data['Total population'].quantile(0.25)
population_q3 = data['Total population'].quantile(0.75)

# calculate the IQR (Inter Quantile Range)
population_IQR = population_q3-population_q1

# Define outlier boundariesa
population_lower_bound = population_q1 - 1.5 * population_IQR
population_upper_bound = population_q3 + 1.5 * population_IQR

# filtering the rows containing outliers
population_outliers = data[(data['Total population'] < population_lower_bound) | (data['Total population'] > population_upper_bound)]
print(len(population_outliers))

population_median = data['Total population'].median()
# dataFrame.loc[condition, 'Column_Name'] = 'Value'

data.loc[population_outliers.index,'Total population']=population_median

# Check if any value still exceeds your ORIGINAL upper bound
population_outliers = data[((data['Total population'] > population_upper_bound) | (data['Total population']< population_lower_bound))]
print(f"Total population outliers : {len(population_outliers)}") 
# This will output: 0
```

```python
# outliers for Total household number
# Calculate Q1 (25th percentile) and Q3 (75th percentile)
household_q1 = data['Total household number'].quantile(0.25)
household_q3 = data['Total household number'].quantile(0.75)

# calculate the IQR (Inter Quantile Range)
household_IQR = household_q3 - household_q1

# Define outlier boundaries
household_lower_bound = household_q1 - 1.5 * household_IQR
household_upper_bound = household_q3 +1.5 * household_IQR

# filtering the rows containing outliers
household_outliers = data[(data['Total household number']< household_lower_bound) | (data['Total household number']>household_upper_bound)]
print(len(household_outliers))
household_median = data['Total household number'].median()
# dataFrame.loc[condition, 'Column_Name'] = 'Value'
data.loc[household_outliers.index,'Total household number']=household_median

# Check if any value still exceeds your ORIGINAL upper bound
household_outliers = data[((data['Total population'] > household_upper_bound) | (data['Total population']< household_lower_bound))]
print(f"household  outliers: {len(household_outliers)}") 
```

18. Visualize Average Family Size distribution across districts using a violin plot.

```python
# 1. Find the top 10 districts by looking at the head(10) names
top_districts_list = data.sort_values(by='Total population', ascending=False).head(10)['District'].unique()

# 2. Filter the ORIGINAL dataset to include ALL rows belonging to those districts
filtered_data = data[data['District'].isin(top_districts_list)]

plt.figure(figsize=(20, 10))

# 3. Plot using the full data subset so there are enough rows to form violins
sns.violinplot(
    data=filtered_data,
    x='District',
    y='Average family size',
    inner='box',
    palette='muted',
    hue='District'
)

plt.title('Average Family Size Distribution within the Top 10 Most Populous Districts', fontsize=14, fontweight='bold')
plt.xticks(rotation=45,fontsize=16)
plt.yticks(fontsize=16)
plt.xlabel("District",fontsize=16)
plt.ylabel("Average family size",fontsize=16)
plt.tight_layout()
plt.show()
```

connecting to the postgresql server

```python
import psycopg2
```

```python
# Connect to PostgreSQL database
def connect_with_psycopg2():
    try:
        # Database connection parameters
        connection = psycopg2.connect(
            host="localhost",        # or your server IP
            database="EDA projects", # replace with your database name
            user="postgres",    # replace with your username
            password="bikash", # replace with your password
            port="5432"              # default PostgreSQL port
        )
    
        print("Connection to PostgreSQL database successful.")
        return connection
    except Exception as e:
        print("Error while connecting to PostgreSQL database:", e)
        return None
# Example usage
connection = connect_with_psycopg2()
```

```python
# Load CSV into PostgreSQL
from sqlalchemy import create_engine
import pandas as pd
def load_csv_to_postgres(csv_file, table_name):
    try:
        # Read the CSV file
        df = pd.read_csv(csv_file)
        
        # Create SQLAlchemy engine from psycopg2 connection
        connection = connect_with_psycopg2()
        if connection is None:
            return
        
        engine = create_engine('postgresql://postgres:bikash@localhost:5432/EDA projects')
        
        # Load DataFrame to PostgreSQL
        df.to_sql(table_name, engine, index=False, if_exists='replace')
        print(f"CSV file {csv_file} loaded into table {table_name} successfully.")
        
        # Close the connection
        connection.close()
    except Exception as e:
        print("Error while loading CSV to PostgreSQL:", e)

# Example usage
csv_file = "census_data_eda.csv"  # Path to your CSV file
table_name = "cleaned_census_data"    # Desired table name in PostgreSQL
load_csv_to_postgres(csv_file, table_name)
```
