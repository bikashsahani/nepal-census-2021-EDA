4. Visualization (6 questions)

    15. Create a bar chart showing Top 10 Districts by Total Population.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data.sort_values(by='Total population', ascending=False).head(10)
```

```python
plt.figure(figsize=(25,10))
sns.barplot(
    data=data.sort_values(by='Total population', ascending=False).head(10),
    x='District',
    y='Total population'
)
plt.xticks(rotation=90,fontsize=14)
plt.xlabel("Top 10 district",fontsize=20)
plt.ylabel("population", fontsize=20)
plt.title("Top 10 District by population", fontsize=20)
```

16. Plot a histogram and boxplot to show the distribution of Total Population across Local Levels.

```python
plt.figure(figsize=(12, 6))

sns.histplot(
    data=data,  # Make sure this matches your DataFrame variable name
    x='Total population',
    bins=15,    # Adjust the number of bins based on your total row count
    kde=True,   # Keeps the smooth density estimation line
    color='skyblue'
)

# Rotate the x-axis numbers if they are tight, though 90 might be overkill for a clean x-axis
plt.xticks(rotation=45) 

plt.title('Distribution of Total Population Across Local Levels', fontsize=14, fontweight='bold')
plt.xlabel('Total Population', fontsize=12)
plt.ylabel('Count (Frequency of Local Levels)', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))

# Fix: Only pass x, remove y. This tells Seaborn to plot a standard 1D histogram.
sns.boxplot(
    data=data, 
    y='Total population', 
    color='lightgreen',
    width=0.4
)

plt.title('Boxplot of Total Population Distribution', fontsize=14, fontweight='bold')
plt.xlabel('Local Levels', fontsize=12)
plt.ylabel('Total Population', fontsize=12)
plt.tight_layout()
plt.show()
```
